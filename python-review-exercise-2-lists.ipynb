# Create a function called 'blank' that checks the string for blank input

```python
# This code block is a function that 
# checks a blank input from the user
def blank(s):
    
    # uses the strip() function to remove spaces from both ends of the string
    # returns true when the string is blank
    if (s.strip() == ''):
        return True
    
    # returns false when the string is not blank
    else:
        return False
```

# Create a function called 'exist' that removes and stores the index of the duplicated strings in a list

```python
# This code block is a function that 
# checks if the list has duplicated strings
def exist(s):
    
    # initializes empty lists lst and loc
    lst = []
    loc = []
    
    # creates a loop
    # that iterates through each item in s
    for x in s:
        
        # if x is not yet in the list
        # it adds it to the list
        if x not in lst:
            lst.append(x)
        
        # if it already exist
        # it adds the index of the item to loc
        else:
            loc.append(s.index(x))
    
    # returns all the list
    return lst, loc
```

# Running and Debugging the Program

```python
# This code block
# runs the code for debugging

# initializes an empty list arr
arr = []

# asks the user for input
# then adds the input to arr
x = input("Please enter anything: ")
arr.append(x)

# creates a while loop that stops when 
# the user has entered blank/spaces
while not blank(x):
    x = input("Please enter anything: ")
    arr.append(x)
    
# adds a new line
print("\n")

# removes the extra blank '' input from the end of the list
arr.pop()

# calls the exist function
# stores the unduplicated list to 'first'
# stores the duplicated indexes in a list 'second'
first, second = exist(arr)

# creates a loop
# that extracts each item in 'first' to 'value'
# then prints each value
for value in first:
    print(f"Output: {value}")
```
